# Perceptron Learning with Stochastic Gradient Descent

```python
%matplotlib inline  

import numpy as np
from random import choice
import matplotlib.pyplot as plt
```

```python
unit_step = lambda x: 0 if x < 0 else 1
unit_step_vector = np.vectorize(unit_step)
```

## Generating Data

```python
X = np.random.rand(500, 2)
X = np.hstack((X, np.ones((500, 1)))) # Add colummns of ones for bias
X_test = np.random.rand(100, 2)
X_test = np.hstack((X_test, np.ones((100, 1)))) # Add colummns of ones for bias

Y = unit_step_vector(X.dot(np.array([2,-3,1])))

n_samples = X.shape[0]
n_features = X.shape[1]
```

### Plotting training dataset

```python
X1 = X[Y.astype(bool),:]
X2 = X[~Y.astype(bool),:]
plt.plot(X1[:,0], X1[:,1], 'ro')
plt.plot(X2[:,0], X2[:,1], 'go')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

## Training

### Initializing weights

```python
weights = np.random.randn(n_features)
print weights
```

### Training with SGD

```python
epochs = 500
eta = 0.1
sample_size = 50

print 'epoch: 0'
# plot decision boundary
plt.plot(X1[:,0], X1[:,1], 'ro')
plt.plot(X2[:,0], X2[:,1], 'go')
plt.plot(X[:,0],-1*(X[:,0]*weights[0] + weights[2])/weights[1], 'b')
plt.xlabel('x1')
plt.ylabel('x2')
plt.axis((0,1,0,1))
plt.show()

for i in range(epochs):
            idx = np.random.choice(len(Y), sample_size)
            x = X[idx]
            y = Y[idx]
            Yp = unit_step_vector(np.matmul(x,weights.T))
            e = np.repeat((Yp - y).T,3,axis=-1).reshape(x.shape)
            dw = (x * e).mean(axis = 0)
            weights -= eta * dw # weight update
            
            if not i%50:
                print 'epoch: ' + str(i)
                # plot decision boundary
                plt.plot(X1[:,0], X1[:,1], 'ro')
                plt.plot(X2[:,0], X2[:,1], 'go')
                plt.plot(X[:,0],-1*(X[:,0]*weights[0] + weights[2])/weights[1], 'b')
                plt.xlabel('x1')
                plt.ylabel('x2')
                plt.axis((0,1,0,1))
                plt.show()

print 'Learned weights: '
print weights

print 'Actual weights: '
print [2,-3,1]
```

## Prediction

```python
Y = unit_step_vector(X_test.dot(np.array([2,-3,1])))
Yp = unit_step_vector(np.matmul(X_test,weights.T))
print(zip(Y,Yp))
```
