# Perceptron Learning with Stochastic Gradient Descent

```python
%matplotlib inline  

import numpy as np
from random import choice
import matplotlib.pyplot as plt
```

```python
unit_step = lambda x: 0 if x < 0 else 1
unit_step_vector = np.vectorize(unit_step)
```

## Generating Data

```python
X = np.random.rand(50, 2)
X = np.hstack((X, np.ones((50, 1)))) # Add colummns of ones for bias
X_test = np.random.rand(10, 2)
X_test = np.hstack((X_test, np.ones((10, 1)))) # Add colummns of ones for bias

Y = unit_step_vector(X.dot(np.array([2,-3,1])))

n_samples = X.shape[0]
n_features = X.shape[1]
```

### Plotting training dataset

```python
X1 = X[Y.astype(bool),:]
X2 = X[~Y.astype(bool),:]
plt.plot(X1[:,0], X1[:,1], 'ro')
plt.plot(X2[:,0], X2[:,1], 'go')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

## Training

### Initializing weights

```python
weights = np.random.randn(n_features)
print weights
```

### Training with SGD

```python
epochs = 250
eta = 0.1

for i in range(epochs):
            x, y = choice(zip(X,Y))
            Yp = unit_step(x.dot(weights)) # predicted output
            dw = (Yp - y) * x
            weights -= eta * dw # weight update
            
            if not i%10:
                print 'epoch: ' + str(i)
                # plot decision boundary
                plt.plot(X1[:,0], X1[:,1], 'ro')
                plt.plot(X2[:,0], X2[:,1], 'go')
                plt.plot(X[:,0],-1*(X[:,0]*weights[0] + weights[2])/weights[1], 'b')
                plt.xlabel('x1')
                plt.ylabel('x2')
                plt.show()

print 'Learned weights: '
print weights
```

## Prediction

```python
for sample in X_test:
    Y_predicted = unit_step(sample.dot(weights))
    print Y_predicted
```
